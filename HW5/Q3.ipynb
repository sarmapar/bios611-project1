```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from sklearn.manifold import TSNE
```

```python
data = pd.read_csv(r'char_stats.csv')
data_numeric = data[["Intelligence","Strength","Speed","Durability","Power","Combat","Total"]]
```

```python
tsne = TSNE(n_components=2)
```

```python
tsne_results = tsne.fit_transform(data_numeric)
data['tsne1'] = tsne_results[:,0]
data['tsne2'] = tsne_results[:,1]
data_tsne = data[["Alignment","tsne1","tsne2"]]
data_tsne
```

```python
data_tsne.to_csv(r'tsne.csv')
```

```python
cols=[]
for i in data_tsne[["Alignment"]].to_numpy().tolist():
    if i==['bad']:
        cols.append('red')
    elif i==['good']:
        cols.append('green')
    elif i==["neutral"]:
        cols.append('blue')
    else:
        cols.append('gray')

plt.scatter(data_tsne[["tsne1"]], data_tsne[["tsne2"]], c=cols)
plt.ylabel("V2")
plt.xlabel("V1")
plt.savefig('tsne_plot_python.png')
```
